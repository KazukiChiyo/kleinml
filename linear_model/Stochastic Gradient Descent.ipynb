# Stochastic Gradient Descent

```python
from SGDRegressor import *
import matplotlib.pyplot as plt

train_x, train_y, test_x, test_y = util.load_lm(slope=2, intercept=1.5, sd=50, n=200)
for e in np.logspace(-6, -3, num=4):
    model = SGDRegressor(batch_size=16, learning_rate="constant", eta0=e)
    model.fit(train_x, train_y)
    score = model.score(test_x, test_y)
    print("eta is: "+str(e))
    print("score is: "+str(score))
    print("")

model = SGDRegressor(batch_size=16, learning_rate="constant", eta0=1e-4)
model.fit(train_x, train_y)
pred = model.predict(test_x)
plt.scatter(train_x, train_y)
plt.plot(test_x, pred, color='red', linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
