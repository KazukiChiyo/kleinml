## PCA
Linear dimensionality reduction using Eigendecomposition of the data to project it to a lower dimensional space.
```
decomposition.PCA(n_components=None)
```

### Parameters
```
n_components: int, or None
    Number of components to keep.
```

### Methods
`fit(X, y)`: Fit the model with X.<br>
`transform(X)`: Apply dimensionality reduction to X.<br>

### Example

````python
import sys
sys.path.append('../')
from decomposition.PCA import PCA
import matplotlib.pyplot as plt
import util

train_x, train_y, test_x, test_y = util.load_diabetes()
model = PCA(n_components=2)
model.fit(train_x)
print("Eigenvectors", model.components)
train_x = model.transform(train_x)
test_x = model.transform(test_x)
plt.scatter(train_x[:, 0], train_x[:, 1], c=train_y, cmap=plt.cm.coolwarm)
````
